```python
import pandas as pd

df_train = pd.read_csv("/content/test_common_voice/test.tsv", sep="\t")
df_train["class"] = 0
df_train["path"] = df_train["path"].apply(lambda x: os.path.join("/content/common_voice/clips", x + ".mp3"))
new_df = df_train[["path", "class"]][:1500]
ones = os.listdir("/content/wakeword_test_data")

data = []
for o in ones:
  data.append([os.path.join("/content/wakeword_test_data", o), 1])

df_wake = pd.DataFrame(data, columns=["path", "class"])
df_wake.head()
df = pd.concat([df_wake, new_df], ignore_index=True)
df = df.sample(frac=1)
print(df.shape)
df.head()
df.reset_index(drop=True)
df.to_csv("test_data_wakeword.csv", index=False)
```
